```python
#Import Libraries, Pandas into Juypter Notebook
import requests
import pandas as pd 
```

```python
# Pull data from API

url = 'https://www.fruityvice.com/api/fruit/all'
r = requests.get(url)
```

```python
# Print data

food = r.json()
print(food)
```

```python
# Print dataframe

df = pd.DataFrame(food)
df
```

```python
# Write a query in Pandas to select a particular set of data - Drop two columns of data
fruits = df.drop(columns=["id","order"])
# View first five rows of data to confirm that the "id" and "order" columns were dropped
fruits.head()
```

```python
# Write a query in Pandas to select a particular set of data - Select only Rosaceae family fruits
fruits[fruits["family"] == "Rosaceae"]
```
