```python
#Import Libraries, Pandas into Juypter Notebook
import requests
import pandas as pd 
```

```python
# Requirement: Pull in data from an API.
# Notes: Import code from library and fetch data. Save the file into local environment. Read the file into a dataframe. 

url = "https://date.nager.at/api/v3/publicholidays/1966/US"
r = requests.get(url)
json = r.json()
```

```python
# Create the dataframe
df = pd.DataFrame(json["data"])
# View first five rows of dataframe
print(df.head())
```

```python
# Write a query in Pandas to select a particular set of data - Drop two columns of data
characters = df.drop(columns=["appearances","id"])
# View first five rows of data to confirm that the "id" and "appearances" columns were dropped
characters.head()
```

```python
# Write a query in Pandas to select a particular set of data - Select only female characters
characters[characters["gender"] == "Female"]
```
