Data Set Description 

Alzheimer's Disease and Healthy Aging Data
U.S. Department of Health & Human Services - 2015-2020. This data set contains data from BRFSS. For reference, BRFSS is the nation's premier system of health-related telephone surveys that collect state data about U.S. residents regarding their health-related risk behaviors, chronic health conditions, and use of preventive services.

```python
import numpy as np
import pandas as pd
Alzheimer_Data = pd.read_csv("Alzheimer_s_Disease_and_Healthy_Aging_Data.csv", low_memory=False)
Alzheimer_Data.head()
```

```python
# Write a query in Pandas to select a particular set of data - Select only state of Florida data
Alzheimer_Data[Alzheimer_Data["LocationDesc"] == "Florida"]
```

```python
# Write a query in Pandas to rename select column names
Alzheimer_Data.rename(columns = {'LocationAbbr':'StateAbbr', 'LocationDesc':'StateName'}, inplace = True)
# View first five rows of data to confirm that the column names were renamed
Alzheimer_Data.head()
```

```python
# Write a query in Pandas to replace NULL values in the "Report" column with the number 0
Alzheimer_Data["Report"].fillna(0, inplace = True)
# Print dataframe
df = pd.DataFrame(Alzheimer_Data)
df
```
